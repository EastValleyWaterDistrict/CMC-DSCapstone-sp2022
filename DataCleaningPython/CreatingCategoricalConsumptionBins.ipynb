# Creating Categorical Variables from the Continous Data 

```python
# Import libraries
import numpy as np      
import pandas as pd    
from IPython.display import display
import matplotlib.pyplot as plt
import random
from scipy.stats.mstats import winsorize
from IPython.display import display
```

```python
# Read in the data 
filename = '../CleanedData/59HydroCleaned.csv'
all_data = pd.read_csv(filename)      
print(f"{filename} : file read into a pandas dataframe.")
```

```python
# Display the Data 
display(all_data)
```

```python
# Get the Max, Min, Mean, and Median 
max_val = all_data['Consumption'].max()
print(max_val)
print(all_data['Consumption'].min())
print(all_data['Consumption'].mean())
print(all_data['Consumption'].median())
```

## Now we cut the Consumption data into categories/bins

```python
def create_consumption_bins(max, bin_size):
    """
    This function creates a list of consumption data bin values
    """
    bins_list = []

    val = 0

    while val <= (max+bin_size):
        bins_list += [val]
        val += bin_size
    
    return bins_list
```

```python
# Create the bins list 
bin_list = create_consumption_bins(max_val, 5000)
bin_list_len = len(bin_list)
print(bin_list_len)
```

```python
classnames = []

for num in range(bin_list_len):
    classnames += [str(num)]

print(classnames)
```

```python
# use Cut to create the consumption data bins

consumption_categories = pd.cut(all_data.Consumption, bins = bin_list, labels = bin_list[1:] )
print(consumption_categories)
all_data.insert(0, "consumption", consumption_categories)

consumption_categories2 = pd.cut(all_data.Consumption_t1, bins = bin_list, labels = bin_list[1:] )
print(consumption_categories)
all_data.insert(0, "t1", consumption_categories)

consumption_categories3 = pd.cut(all_data.Consumption_t2, bins = bin_list, labels = bin_list[1:] )
print(consumption_categories)
all_data.insert(0, "t2", consumption_categories)

consumption_categories4 = pd.cut(all_data.Consumption_t7, bins = bin_list, labels = bin_list[1:] )
print(consumption_categories)
all_data.insert(0, "t7", consumption_categories)

consumption_categories5 = pd.cut(all_data.Consumption_t14, bins = bin_list, labels = bin_list[1:] )
print(consumption_categories)
all_data.insert(0, "t14", consumption_categories)

consumption_categories6 = pd.cut(all_data.Consumption_t21, bins = bin_list, labels = bin_list[1:] )
print(consumption_categories)
all_data.insert(0, "t21", consumption_categories)

```

```python
display(all_data)
```

```python

all_data = all_data.drop(["Consumption", "Consumption_t1", "Consumption_t2", "Consumption_t7", "Consumption_t14", "Consumption_t21"], axis = 1)
display(all_data)
```

```python
# Move the consumption column to the end of the data frame 
df1 = all_data.pop('consumption') 
all_data['consumption']= df1 

display(all_data)
```
